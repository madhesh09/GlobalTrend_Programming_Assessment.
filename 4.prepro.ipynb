```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('carprices.csv')
df
```

```python
dummy=pd.get_dummies(df['Car Model'])
dummy
```

```python
merged=pd.concat([df,dummy],axis='columns')
merged
```

```python
final=merged.drop(['Car Model','Mercedez Benz C class'],axis='columns')
final
```

```python
from sklearn import linear_model
meg=linear_model.LinearRegression()
x=final.drop(['Sell Price($)'],axis='columns')
x
```

```python
y = final['Sell Price($)']
y
```

```python
meg.fit(x,y)
```

```python
meg.predict([[6900,6,0,1]])
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
dfle=df
dfle['Car Model']=le.fit_transform(df['Car Model'])
dfle
```

```python
x=dfle[['Car Model','Mileage','Age(yrs)']]
x
```

```python
y=dfle[['Sell Price($)']]
y
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
tc=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[0])],remainder='passthrough')
x=tc.fit_transform(x)
x
```

```python
x=x[:,1:]
x
```

```python
meg.fit(x,y)
```

```python
meg.predict([[1,0,6900,6]])
```

```python
meg.score(x,y)
```
